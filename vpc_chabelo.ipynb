```python
## tenemos tres puertas, 1 tiene un precio y el participante escoge 1
## De esas 3 chabelo escoge 1 puerta (que obvio no tiene el premio) 
##y el participante tiene las otras dos y escoge si quedarse con la que ya 
#escogió o cambiarlas

#corremos la simulación 2000 veces
puertas=[1,2,3]

# la primera puerta puede ser cualquier de la puerta las 3 puertas
eleccion_1=[random.randint(1,3) for i in range(2000)]

# el premio puede estar en cualquiera de las 3 puertas
premio=[random.randint(1,3) for i in range(2000)]

# chabelo puede escoger cualquier de las 3 puertas
#chabelo=[random.randint(1,3) for i in range(2000)]
```

```python
for i in range(2000):
    if(eleccion_1[i]==puertas[0]):
        index=0
    elif (eleccion_1[i]==puertas[1]):
        index=1       
    else:
        index=2
    print (index)     
```

```python
cambios_puerta=0
no_cambios_puerta=0
eleccion_2=[]
chabelo=[]

for i in range(2000):
    #la puerta que escoge chabelo no es la que tiene el precio
    #o la puerta que escoge el participante 
    
    #la puerta de chabelo es la puerta restante ya que yo escogí 
    #y ya que él sabe cuál tiene el premio
    for puerta_chabelo in range(1,4):
        if(eleccion_1[i] != puerta_chabelo and premio[i] != puerta_chabelo):
            chabelo.append(puerta_chabelo)
    
    if(eleccion_1[i]==puertas[0]):
        index=0
    elif (eleccion_1[i]==puertas[1]):
        index=1       
    else:
        index=2
        
    
    # quito la puerta que yo escogí
    puertas=np.delete(puertas,index)
    
    #si decido no cambiar y gano
    if (eleccion_1[i]==premio[i]): #ya escogió una y es la ganadora
        #gana y no es necesario hacer el cambio
        no_cambios_puerta=no_cambios_puerta+1
        eleccion_2.append(5) #numero al azar,solo para llenar el arreglo
    
    #si sí decido cambiar
    else:
        #solo quedan dos puertas disponibles
        random_choice=random.randint(0,1) # de las puertas disponibles escojo alguna
        eleccion_2.append(puertas[random_choice])
        
        #si la puerta que escogí sí tiene el premio
        if(eleccion_2[i]==premio[i]):
            cambios_puerta=cambios_puerta+1
    
    puertas=[1,2,3] #dejamos el arreglo de puertas igual para la siguiente iteración
        
        
        
        
```
